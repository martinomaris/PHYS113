```python
import matplotlib.pyplot as plt
from scipy import mat, cos, sin, arctan, sqrt, pi, arctan2
from math import pow, degrees, radians
import numpy as np
```

```python
%matplotlib inline
t = np.arange(0.0, 5.0, 0.01)
c = 1.

lam1 = 1.
phi1 = 0.
s1 = np.sin(2*np.pi*c/lam1*t + phi1)

lam2 = 1.01
phi2 = np.pi

s2 = np.sin(2*np.pi*c/lam2*t + phi2)

s3 = s1 + s2

line, = plt.plot(t, s1, lw=2, color='r')
line, = plt.plot(t, s2, lw=2, color='b')
line, = plt.plot(t, s3, lw=4, color='g')
```
