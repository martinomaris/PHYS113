```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 50 * np.pi, 1000)
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1, 1, 1)
line1, = ax.plot(t, np.sin(t), lw=1, color='r')
line2, = ax.plot(t, np.sin(t), lw=1, color='b')
line3, = ax.plot(t, 2*np.sin(t), lw=4, color='g')

def update(w1 = 1.0, q1 = 0.0, w2= 1.0, q2 = 0.0):
    line1.set_ydata(np.sin(w1*t + q1*2.*np.pi))
    line2.set_ydata(np.sin(w2*t + q2*2.*np.pi))
    line3.set_ydata(np.sin(w1*t + q1*2.*np.pi) + np.sin(w2*t + q2*2.*np.pi))
    fig.canvas.draw()

interact(update);
```
